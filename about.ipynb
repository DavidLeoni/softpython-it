# Descrizione del corso

## Introduzione

Al giorno d’oggi, sempre più decisioni vengono prese in base a dati fattuali e oggettivi. Tutte le discipline, dall’ingegneria alle scienze sociali, necessitano oramai capacità di elaborare dati ed estrarre informazioni utili dall’analisi di fonti eterogenee. Questo corso di 9 laboratori pratici fornisce quindi un’introduzione al processamento dati usando [Python](http://www.python.it/), un linguaggio di programmazione popolare sia nell’industria che nell'ambito della ricerca. Gli studenti saranno seguiti dai docenti David Leoni, Marco Caresia e Alessio Zamboni.

## A chi è rivolto 

Il partecipante ideale ha già qualche nozione base di programmazione e desidera acquisire conoscenze pratiche riguardo l’estrazione, la pulizia, l’analisi e visualizzazione di dati (tra i framework Python usati vi saranno Pandas, Numpy e l’editor Jupyter). Per superare eventuali difficoltà e garantire risultati didattici concreti, le guide presentate saranno passo-passo.

## Requisiti

Un computer portatile e qualche conoscenza base di programmazione, in un qualsiasi linguaggio.

## Programma


I materiali presentati durante il corso saranno organizzati in guide passo-passo che saranno pubblicate gratuitamente online all'indirizzo [http://softpython.readthedocs.io](http://softpython.readthedocs.io)


1. Introduzione a Python
    * Installazione, nozioni base su Python (sintassi, programmazione a oggetti)
    * Condivisione script online
2. Formati dei dati
    * Presentazione di formati dati comuni (CSV, JSON, XML, file binari)
    * Convertire dati, trattare errori e dati mal-formattati.
3. Visualizzazione dati
    * Immissione comandi in Jupyter Notebook, plotting di grafici in Matplotlib.
    * Esporre dati online con DataWrapper, RawGraphs, UMap
4. Ricerca
    * Ricerca usando espressioni regolari
    * Ricerca su alberi con CSS
5. Integrazione dati
    * Esempio di integrazione dati: scaricamento di un dataset opendata (i.e. Trentino agritours)
    * pulizia e posizionamento dei punti di interesse su OpenStreetMap usando un servizio di georeferenziazione
6. Estrazione Dati
    * Estrazione di testo rilevante da una pagina HTML (es. eventi del Trentino) usando BeautifulSoup, tagging automatico di testo con la libreria di processamento del testo naturale NLTK
7. Analitiche sui dati
    * Analisi di un dataset contenente dati numerici (es. statistiche sulle alluvioni) con la libreria Pandas. Calcolo di correlazioni tra fattori
8. Applicazioni interattive
    * Creazione di interfacce per analisi interattiva di dati con Jupyter Widgets
    * Discussione di alternative (QT, wxWidget)
9. Integrazione con database
    * Estrazione di dati da un database SQLite con Pandas
    * Esecuzione di semplici ricerche SQL
